# Popularity based Recommendations :

```python
# counting the total ratings

rating_df = blog_df.groupby('blog_title').count()['ratings'].reset_index()
rating_df.rename(columns={'ratings' : 'total_ratings'}, inplace=True)
rating_df
```

```python
blog_df.groupby('blog_title').mean('ratings')
```

```python
# finding the mean of the ratings

avg_rating_df = blog_df.groupby('blog_title').mean('ratings').reset_index()
avg_rating_df.rename(columns={'ratings' : 'avg_ratings'}, inplace=True)
avg_rating_df
```

```python
popular_df = rating_df.merge(avg_rating_df, on='blog_title')[['blog_title', 'total_ratings', 'avg_ratings']]
popular_df
```

```python
popular_df[popular_df['total_ratings']>= 57].sort_values('avg_ratings', ascending=False).shape
```

```python
# top blogs those ratings is 57 and more than 57

top_40_df = popular_df[popular_df['total_ratings']>= 57].sort_values('avg_ratings', ascending=False)
```

```python
top_40_df = top_40_df.merge(blog_df, on='blog_title').drop_duplicates('blog_title')[['blog_title','blog_content','blog_link','blog_img','topic','author_name','total_ratings','avg_ratings']]
```

# There are 5 top rated blogs from each topic in the "top_of_each" data frame

```python
topics = ['ai', 'blockchain', 'cybersecurity', 'web-development', 'data-analysis', 'cloud-computing', 'security', 'web3',
         'machine-learning', 'nlp', 'data-science', 'deep-learning', 'android', 'dev-ops', 'information-security', 'image-processing', 
          'flutter', 'backend', 'cloud-services', 'Cryptocurrency', 'app-development', 'backend-development', 'Software-Development']

data_frames = []
l = [1,2,3,4,5]

for i in topics:
    # Filter blog_df for the current topic
    new_df = blog_df[blog_df['topic'] == i].sort_values('ratings', ascending=False).drop_duplicates('blog_title')[['blog_title','blog_content','blog_link','blog_img','topic','author_name','ratings']]
    # Append the DataFrame to the list
    data_frames.append(new_df[0: len(l)])

#top_of_each = data_frames
top_of_each = pd.concat(data_frames, ignore_index=True)
top_of_each
```

# Collaborative filtering based recommender system

```python
collab_df = blog_df.groupby('userId').count()['ratings'] > 55
freq_reader_df = collab_df[collab_df].index
```

```python
filter_reader_df = blog_df[blog_df['userId'].isin(freq_reader_df)]
filter_reader_df
```

```python
filter_blog_df = filter_reader_df.groupby('blog_title').count()['ratings'] >= 30
famous_blogs = filter_blog_df[filter_blog_df].index
```

```python
final_df = filter_reader_df[filter_reader_df['blog_title'].isin(famous_blogs)]
final_df
```

```python
final_df.drop_duplicates()
```

```python
pivot_table = final_df.pivot_table(index='blog_title', columns='userId', values='ratings')
pivot_table.fillna(0, inplace=True)
pivot_table
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim_score = cosine_similarity(pivot_table)
```

```python
sim_score.shape
```

```python
def blog_recommender(blog_title):
    index = np.where(pivot_table.index == blog_title)[0][0]
    similar = sorted(list(enumerate(sim_score[index])), key= lambda x:x[1], reverse=True)[1:11]
    
    for i in similar:
        print(pivot_table.index[i[0]])
```
